```python
items = browser.find_elements_by_class_name('resultItem')
```

```python
data = []
for item in items:
    title = item.find_element_by_class_name('truncatedResultsTitle').text
    titleAuthorETC0 = item.find_elements_by_class_name('titleAuthorETC')[0].text
    titleAuthorETC1 = item.find_elements_by_class_name('titleAuthorETC')[1].text
    data.append((title, titleAuthorETC0, titleAuthorETC1))
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords = stopwords.words('english')
```

```python
newstopwords = ["china","kong","''china's","world","china's","amid","Chinese","china's","us","'china"]
stopwords.extend(newstopwords)
stopwords
```

```python
removestopwords=[]
for i in splitwords:
    i = i.lower() 
    if i not in stopwords:
        removestopwords.append(i)
removestopwords
```

```python
for i in '':#
            removestopwords = [item.replace(i," ") for item in removestopwords]
removestopwords
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
def word_count(processed_word_list):
    word_count = pd.Series(processed_word_list).value_counts().sort_values(ascending=False)[0:25]  
    fig = plt.figure(figsize=(16,8))  
    x = word_count.index.tolist()  
    y = word_count.values.tolist()  
    sns.barplot(x, y, palette="BuPu_r")  
    plt.title('people daily word frequency top 25')  
    plt.ylabel('count')  
    sns.despine(bottom=True)  
    #plt.savefig('./word_count_bar.png',dpi=400)  
    plt.show()

word_count(removestopwords)
```

```python
from PIL import Image
import wordcloud
import numpy as np

def tag_cloud(text):
    mask = np.array(Image.open('picture.jpg')) #set mask, you can change to the picture you like, but it must have a high color contrast
    wc = wordcloud.WordCloud(mode='RGBA',background_color='white',max_words=2000,stopwords=stopwords,max_font_size=300,random_state=42,mask=mask)
    wc.generate_from_text(' '.join(text))
    plt.figure(figsize=(12,12))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
    plt.title('People Daily, 20170801', loc='Center', fontsize=20)
    plt.show()
    return plt.show()

tag_cloud(removestopwords)
```
