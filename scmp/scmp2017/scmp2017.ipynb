```python
import pandas
df1=pandas.read_csv("scmpxw2017a.csv").sort_values(by="dates",ascending=True)
df2=pandas.read_csv("scmpxw2017b.csv").sort_values(by="dates",ascending=True)
df3=pandas.read_csv("scmpxw2017c.csv").sort_values(by="dates",ascending=True)
df4=pandas.read_csv("scmpxw2017d.csv").sort_values(by="dates",ascending=True)
df5=pandas.read_csv("scmpxw2017e.csv").sort_values(by="dates",ascending=True)
df6=pandas.read_csv("scmpxw2017f.csv").sort_values(by="dates",ascending=True)
df7=pandas.read_csv("scmpxw2017g.csv").sort_values(by="dates",ascending=True)
```

```python
df2017=pandas.concat([df1,df2,df3,df4,df5,df6,df7], ignore_index=True)
```

```python
df2017

del df2017["Unnamed: 0"]

from datetime import datetime
from dateutil import parser
import numpy
def parse_datetime(x):
    try:
        return parser.parse(x)
    except:
        return numpy.nan
df2017['datetime'] = df2017['dates'].apply(parse_datetime)
df2017['datetime']
df2017_order=df2017.set_index('datetime').resample('1d').aggregate('count')
df2017_order

dfc=df2017_order["title"]==0
df2017_order[dfc]
```

```python
df2017['datetime'] = df2017['dates'].apply(parse_datetime)
df2017new=df2017.sort_values(by='datetime')
df2017new.dropna(subset=['title'], how='all', inplace = True)
df2017new
```

```python
df2017new.to_csv("scmp2017.csv",encoding='utf-8')
```
