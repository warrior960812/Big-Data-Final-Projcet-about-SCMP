```python
import pandas
df1=pandas.read_csv("scmp201504.csv").sort_values(by="dates",ascending=True)
df2=pandas.read_csv("scmp201505.csv").sort_values(by="dates",ascending=True)
df3=pandas.read_csv("scmp2015.csv").sort_values(by="dates",ascending=True)
df4=pandas.read_csv("scmp201509.csv").sort_values(by="dates",ascending=True)
df5=pandas.read_csv("scmp201512.csv").sort_values(by="dates",ascending=True)
df6=pandas.read_csv("scmpxw2015a.csv").sort_values(by="dates",ascending=True)
```

```python
df2015=pandas.concat([df1,df2,df3,df4,df5,df6], ignore_index=True)
```

```python
df2015
del df2015["Unnamed: 0"]
```

```python
df2015
```

```python
from datetime import datetime
from dateutil import parser
import numpy
def parse_datetime(x):
    try:
        return parser.parse(x)
    except:
        return numpy.nan
df2015['datetime'] = df2015['dates'].apply(parse_datetime)
df2015['datetime']
df2015_order=df2015.set_index('datetime').resample('1d').aggregate('count')
df2015_order
```

```python
dfc=df2015_order["title"]==0
df2015_order[dfc]
```

```python
df2015['datetime'] = df2015['dates'].apply(parse_datetime)
df2015new=df2015.sort_values(by='datetime')
df2015new
```

```python
df2015new.dropna(subset=['title'], how='all', inplace = True)
df2015new
```

```python
df2015new.to_csv("scmp2015.csv",encoding='utf-8')
```
