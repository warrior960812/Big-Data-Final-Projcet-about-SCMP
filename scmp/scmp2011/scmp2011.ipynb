```python
import pandas
```

```python
df1=pandas.read_csv("scmp2011.csv").sort_values(by="dates",ascending=True)
df2=pandas.read_csv("scmp2011_add1.csv").sort_values(by="dates",ascending=True)
df3=pandas.read_csv("scmp2011_add2.csv").sort_values(by="dates",ascending=True)
df4=pandas.read_csv("scmp2011_add3.csv").sort_values(by="dates",ascending=True)
```

```python
df2011=pandas.concat([df1,df2,df3,df4])
```

```python
del df2011["Unnamed: 0"]
```

```python
df2011
```

```python
from datetime import datetime
from dateutil import parser
import numpy
def parse_datetime(x):
    try:
        return parser.parse(x)
    except:
        return numpy.nan
df2011['datetime'] = df2011['dates'].apply(parse_datetime)
df2011new=df2011.sort_values(by='datetime')
df2011new
```

```python
df2011new.to_csv("scmp2011.csv",encoding='utf-8')
```
