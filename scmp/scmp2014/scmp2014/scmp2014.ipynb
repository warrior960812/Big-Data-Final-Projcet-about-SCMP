```python
import pandas
```

```python
df1=pandas.read_csv("scmp201401-201402.csv").sort_values(by="dates",ascending=True)
df2=pandas.read_csv("scmp201403-201404.csv").sort_values(by="dates",ascending=True)
df3=pandas.read_csv("scmp201405-201408.csv").sort_values(by="dates",ascending=True)
df4=pandas.read_csv("scmp201409-201410.csv").sort_values(by="dates",ascending=True)
df5=pandas.read_csv("scmp201411-201412.csv").sort_values(by="dates",ascending=True)
df6=pandas.read_csv("scmp2014_add1.csv").sort_values(by="dates",ascending=True)
df7=pandas.read_csv("scmp2014_add2.csv").sort_values(by="dates",ascending=True)
df8=pandas.read_csv("scmp2014_add3.csv").sort_values(by="dates",ascending=True)
df9=pandas.read_csv("scmp2014_add4.csv").sort_values(by="dates",ascending=True)
df10=pandas.read_csv("scmp2014_add5.csv").sort_values(by="dates",ascending=True)
df11=pandas.read_csv("scmp2014_add6.csv").sort_values(by="dates",ascending=True)
```

```python
df2014=pandas.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11])
```

```python
del df2014["Unnamed: 0"]
```

```python
df2014
```

```python
from datetime import datetime
from dateutil import parser
import numpy
def parse_datetime(x):
    try:
        return parser.parse(x)
    except:
        return numpy.nan
df2014['datetime'] = df2014['dates'].apply(parse_datetime)
df2014['datetime']
df2014_order=df2014.set_index('datetime').resample('1d').aggregate('count')
df2014_order
```
