```python
import pandas
```

```python
df7=pandas.read_csv("scmp2012.csv").sort_values(by="dates",ascending=True)
df8=pandas.read_csv("scmp201202.csv").sort_values(by="dates",ascending=True)
df9=pandas.read_csv("scmp201203.csv").sort_values(by="dates",ascending=True)
df10=pandas.read_csv("scmp201204.csv").sort_values(by="dates",ascending=True)
df11=pandas.read_csv("scmp201206.csv").sort_values(by="dates",ascending=True)
df12=pandas.read_csv("scmp2012aa.csv").sort_values(by="dates",ascending=True)
df13=pandas.read_csv("scmp2012bbcsv.csv").sort_values(by="dates",ascending=True)
df14=pandas.read_csv("scmp2012cc.csv").sort_values(by="dates",ascending=True)
```

```python
df2012=pandas.concat([df7,df8,df9,df10,df11,df12,df13,df14], ignore_index=True,sort=False)
df2012
del df2012["Unnamed: 0"]
```

```python
from datetime import datetime
from dateutil import parser
import numpy
def parse_datetime(x):
    try:
        return parser.parse(x)
    except:
        return numpy.nan
df2012['datetime'] = df2012['dates'].apply(parse_datetime)
df2012['datetime']
df2012_order=df2012.set_index('datetime').resample('1d').aggregate('count')
df2012_order

df2012
```
