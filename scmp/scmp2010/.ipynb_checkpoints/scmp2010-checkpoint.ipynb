```python
import pandas
```

```python
df1=pandas.read_csv("scmp2010a.csv").sort_values(by="dates",ascending=True)
df2=pandas.read_csv("scmp2010b.csv").sort_values(by="dates",ascending=True)
df3=pandas.read_csv("scmp2010c.csv").sort_values(by="dates",ascending=True)
df4=pandas.read_csv("scmp2010d.csv").sort_values(by="dates",ascending=True)
df5=pandas.read_csv("scmp2010e.csv").sort_values(by="dates",ascending=True)
df6=pandas.read_csv("scmp2010f.csv").sort_values(by="dates",ascending=True)
df7=pandas.read_csv("scmp2010g.csv").sort_values(by="dates",ascending=True)
df8=pandas.read_csv("scmp2010h.csv").sort_values(by="dates",ascending=True)
```

```python
df2010=pandas.concat([df1,df2,df3,df4,df5,df6,df7,df8], ignore_index=True)
```

```python
df2010
```

```python
#把日期照顺序排列# 
from datetime import datetime
from dateutil import parser
import numpy
def parse_datetime(x):
    try:
        return parser.parse(x)
    except:
        return numpy.nan
df2010['datetime'] = df2010['dates'].apply(parse_datetime)
df2010['datetime']
df2010_order=df2010.set_index('datetime').resample('1d').aggregate('count')
df2010_order
dfc=df2010_order["title"]==0
df2010_order[dfc]
```

```python
df2010['datetime'] = df2010['dates'].apply(parse_datetime)
df2010new=df2010.sort_values(by='datetime')
df2010new
```

```python
df2010.to_csv("scmp2010.csv",encoding='utf-8')
```

```python
def China(t):
    return 'China' in str(t).lower()
def Ma(t):
    return 'ma' in str(t).lower()
df2010['kw-China'] = df2010['title'].apply(China)
df2010['kw-Ma'] = df2010['title'].apply(Ma)
df2010.set_index('datetime').resample('1d').aggregate('sum').tail()
```
