```python
import csv
import pandas as pd
```

# 分析标题的关键词

```python
with open('scmp2018.csv','r') as f: 
    rows = csv.reader(f) 
    title_list = [row[1] for row in rows]
a = str(title_list)
title_split = a.split()
#  title_split
```

```python
def stopwordslist(filepath):   
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]  
    return stopwords
```

```python
stopwords = stopwordslist('stopwords.txt')
```

```python
title_new_list=[]
for i in title_split:
    i = i.lower() 
    if i not in stopwords:
        title_new_list.append(i)
title_new_list
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
def word_count(title_new_list):
    word_count = pd.Series(title_new_list).value_counts().sort_values(ascending=False)[0:20]  
    fig = plt.figure(figsize=(18,10))  
    x = word_count.index.tolist()  
    y = word_count.values.tolist()  
    sns.barplot(x, y, palette="BuPu_r")  
    plt.title('reporter publication top 5')  
    plt.ylabel('count')  
    sns.despine(bottom=True)  
    #plt.savefig('./word_count_bar.png',dpi=400)  
    plt.show()

word_count(title_new_list) 
```
