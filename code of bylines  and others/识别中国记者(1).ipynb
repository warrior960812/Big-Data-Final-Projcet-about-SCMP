```python
import csv
import pandas as pd
```

```python
pinyin_Lib = open('baijiaxing.txt', 'r')
a = pinyin_Lib.read()
all_surname_list = a.split(',')
```

```python
Chinese_reporter_rate=[]
for i in range(2010,2019):   
    file_name = 'scmp' + str(i) + ".csv"
    with open(file_name,'r') as f: 
        rows = csv.reader(f) 
        byline_list = [row[2] for row in rows]
    b = str(byline_list)
    c = b.replace("''","").replace(".","")
    d = c.split("ï¼›")
    list1= str(d).split(",")
    str2 = str(list1).lower()
    name_list = str2.split()
    result = 0
    author_count=b.count(",")
    for n in range(len(name_list)):
        if name_list[n] in all_surname_list:
            result = result + 1
    x = result/author_count
    rate = round(x,5)
    Chinese_reporter_rate.append(rate)
```

```python
Chinese_reporter_rate
```

```python
year=['2010','2011','2012','2013','2014','2015','2016','2017','2018']
```

```python
list3 = []
for o in range(0,7):
    row = (year[o],Chinese_reporter_rate[o])
    list3.append(row)
print(list3)
import csv
with open('Chinese.csv','w') as f:
    mywriter=csv.writer(f)
    mywriter.writerows(list3)
```
