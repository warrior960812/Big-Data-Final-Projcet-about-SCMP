```python
import pandas as pd
import csv
```

```python
Ma=[]
Alibaba=[]
Amazon=[]
title=[]
Tencent=[]
year=['2010','2011','2012','2013','2014','2015','2016','2017','2018']

for i in range(2010,2019):
    file_name = 'scmp' + str(i) + '.csv'
    with open(file_name,'r') as f: 
        rows = csv.reader(f) 
        n = [row[1] for row in rows]
    a = str(n)
    str2 = a.replace("Jack Ma","AAAA",300)
    m = str2.split()
    count_Ma = m.count('AAAA')
    Ma.append(count_Ma)
    count_Alibaba = m.count('Alibaba')
    Alibaba.append(count_Alibaba)
    count_Amazon = m.count('Amazon')
    Amazon.append(count_Amazon)
    count_Tencent = m.count('Tencent')
    Tencent.append(count_Tencent)
    title_len = len(m)
    title.append(title_len)
    
```

```python
index = [year]
df = pd.DataFrame({'Alibaba':Alibaba,'Amazon':Amazon,'Tencent':Tencent}, index=index)
ax = df.plot.bar(rot=10)
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10, 10)
plt.title('Company Frequency', color='gray', fontsize=25, weight='bold')
plt.xticks(fontsize=15)
plt.yticks(fontsize=20)
ax.legend(fontsize=20)
```
